```python
from fastai.vision.all import *

# 1. Load Data
path = Path('../data')
dls = ImageDataLoaders.from_folder(path, train='train', valid='test', item_tfms=Resize(224))

# 2. Train
learn = vision_learner(dls, resnet50, metrics=accuracy)
learn.fine_tune(5)

# 3. Save the Brain
learn.export('../models/model_export.pkl')
```
